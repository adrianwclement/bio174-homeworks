__Exercise 5:__ Use the yeast genome gff file that we downloaded in class.

Read about the gff format and write a script that will create a database table with all the yeast genes and chromosomes. GFF files have all the annotation for a genome. For simplicity in our table we are only interested in the lines of the GFF files that give the coordinates of the chromosomes, genes, and tRNA genes. Your table should have the following columns: `ID number, Chromosome number, Start Position, End Position, Strand, Feature, Description`.

All this information should be readily available from the gff file, with the exception of the ID number, which is given as the "ID" in the last column, and you will have to process it a little bit. Again, only include in your table the yeast protein coding genes, tRNA coding genes, and chromosomes. Your final table should have one row per gene for a total of 6,923 rows.

Note that the description of several genes, and even some gene names contain special HTML characters, like %20. To convert these to text you should use the `unquote()` function of the `urllib` package. Do that before trying to use your regular expressions.

```python
from urllib.parse import unquote
import sqlite3 as sql


# Parses the ID and Description from the attributes column
def parse_attributes(attributes):
    attribute_dict = {}
    for attribute in attributes.split(';'):
        if '=' in attribute:
            key, value = attribute.split('=')
            attribute_dict[key] = value

    id_number = attribute_dict.get('ID', '')
    description = unquote(attribute_dict.get('Note', ''))

    return id_number, description


# Load the GFF file, skipping comments
gff_path = './saccharomyces_cerevisiae_R64-3-1_20210421.gff'
gff_data = pd.read_csv(
    gff_path,
    sep='\t',
    comment='#',
    header=None,
    names=['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute'],
    dtype={'start': str, 'end': str, 'score': str, 'strand': str, 'frame': str},
    low_memory=False
)


# Filter for desired features
features_of_interest = ['gene', 'tRNA_gene', 'chromosome']
filtered_data = gff_data[gff_data['feature'].isin(features_of_interest)].copy()


# Parse the ID and Description
attributes_parsed = filtered_data['attribute'].apply(parse_attributes)
filtered_data.loc[:, 'ID Number'] = attributes_parsed.apply(lambda x: x[0])
filtered_data.loc[:, 'Description'] = attributes_parsed.apply(lambda x: x[1])


# Create a new DataFrame with the specific columns
final_data = filtered_data[['ID Number', 'seqname', 'start', 'end', 'strand', 'feature', 'Description']]
final_data.columns = ['ID Number', 'Chromosome Number', 'Start Position', 'End Position', 'Strand', 'Feature', 'Description']


# Reset the index of the DataFrame, drop the old index, and ensure all numbers are of type int
final_data.reset_index(drop=True, inplace=True)
final_data.loc[:, 'Start Position'] = final_data['Start Position'].astype(int)
final_data.loc[:, 'End Position'] = final_data['End Position'].astype(int)
# final_data


# Connect to SQLite and write the DataFrame to a new table
con = sql.connect('orthologs.sqlite')
final_data.to_sql('gff_new', con, if_exists='replace', index=False)


# Confirm completion
print("Database table created successfully with", len(final_data), "rows.")
```
